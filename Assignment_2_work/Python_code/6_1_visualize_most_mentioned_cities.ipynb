# Visualize the most mentioned Ukrainian cities
This will give insight into what areas have the most talked about conflicts

```python
word_frequency_list = []
reduced_ordered_text_file = file_path = "/Users/hannojacobs/MIT805_datasets/mapreduced_tweet_fields/post_processed_tweet_fields/ordered_text_words.txt"

with open(file_path, "r") as file:
    for line in file:
        parts = line.strip().split()
        if len(parts) == 2:
            word = parts[0]
            frequency = int(parts[1])
            word_frequency_list.append([word, frequency])

print(word_frequency_list[0:100])
```

# Now get out the values that have cities named

```python
cities_list = [
    "Kyiv",
    "Kharkiv",
    "Odessa",
    "Dnipro",
    "Lviv",
    "Donetsk",
    "Zaporizhzhia",
    "KryvyiRih",
    "Ternopil",
    "Chernihiv",
    "Kherson",
    "Poltava",
    "Cherkasy",
    "IvanoFrankivsk",
    "Kropyvnytskyi",
    "Sumy"
] # all of the spaces and '-' characters have been removed

named_cities_list = []
for city in cities_list:
    city_hashtag = city.lower()
    
    for word, frequency in word_frequency_list:
        if city_hashtag in word:
            named_cities_list.append([word, frequency])

# Sort the named_cities_list
named_cities_list = sorted(named_cities_list, key=lambda x: x[1], reverse=True)


# Print the list of city name and frequency pairs
print(named_cities_list)
```

## Plot these in a histogram

```python
import matplotlib.pyplot as plt

num_words_to_take = 40
top_X_words_and_freq = named_cities_list[0:num_words_to_take-1]
top_X_words, top_X_freq = zip(*top_X_words_and_freq)

plt.figure(figsize=(12, 10))  # Adjust the figure size as needed
plt.barh(top_X_words, top_X_freq, color='skyblue')
plt.gca().invert_yaxis()  # Reverse the order to have the most common word at the top
plt.xlabel('Frequency')
plt.ylabel('Locations')
plt.title(f'Top {num_words_to_take} Location hashtags by Frequency')
plt.tight_layout()
plt.show()
```

## Takeaway

* From this we can see that Kyiv and Kharhiv are the most common cities named in a hashtag.
* These hashtags show the most common hashtag's related to cities
